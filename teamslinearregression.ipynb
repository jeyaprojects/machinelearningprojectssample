```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression#(custom dataset)
```

```python
players=pd.read_csv("teams.csv")
```

```python
players
```

```python
players.columns
```

```python
#plot a graph
plt.figure(figsize=(16,5))
plt.scatter(players["athletes"],players["medals"],c="blue")
plt.xlabel("athletes")
plt.ylabel("medals")
plt.show()
```

```python
#build the model
X=players["athletes"].values.reshape(-1,1)
Y=players["medals"].values.reshape(-1,1)
```

```python
#generate model
reg=LinearRegression()
reg.fit(X,Y)
```

```python
#printcoefficient and intercept
print(reg.coef_[0][0])
print(reg.intercept_[0])
```

```python
predictions=reg.predict(X)
```

```python
#graph
plt.figure(figsize=(16,8))
plt.scatter(players["athletes"],players["medals"],c="black")
plt.plot(players["athletes"],predictions,c="blue")
plt.xlabel("age")
plt.ylabel("medals")
plt.show()
```
