```python
Raw_Housing_Data["No of Times Visited"].unique()#it has change after mapping before it was once,none,thrice,twice,four
```

```python
#now lets look at "renovated house " vaiable(is says whether the house is renovated or not).this will impact the sale price(newly renovated house will have high price than old one)
```

```python
#here we need to derive two variables from "Renovated House"
#1-->"ever renovated"-->we list the renovated house
#2-->"year since renovated"
```

```python
#"Ever renovated"-->assign "yes"(if it is renovted) and "No" for "renovate year " is zero-->this is numerical variable
#this is called "categorical variables"{after derive this 2 variable we need to add them in our dataset}
```

```python
#we use where() of numpylibrary
```

```python
Raw_Housing_Data["Ever Renovate"]=np.where(Raw_Housing_Data["Renovated Year"] == 0 ,'No','Yes')
```

```python
Raw_Housing_Data.head()
```

```python
#for second variable-->"Year Since Renovation"="year of sale" -"Year of Renovation"
#"Year of Sale"-->we get it from when house was sold
#for the house which is never renovated ,"Year since Renovation"-->becomes zero(0)
#"Year Since Renovation " is numerical  variables
```

```python
#we need to fetch "year" from "Date house was sold" and name it as"purchase year
```

```python
Raw_Housing_Data["Purchase Year"]=pd.DatetimeIndex(Raw_Housing_Data['Date House was Sold']).year
```

```python
#"Year Since Renovation"="Year of Sale"-"Year of renovation"
```

```python
Raw_Housing_Data["Years Since Renovation"]=np.where(Raw_Housing_Data['Ever Renovated']=='Yes',abs(Raw_Housing_Data['Purchase Year']-Raw_Housing_Data['Renovated Year']),0)
```
